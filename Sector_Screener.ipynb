```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go

# Fetch S&P 500 list from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500_table = pd.read_html(url, header=0)
sp500_df = sp500_table[0]

# Function to fetch and plot data for a given sector
def plot_sector_data(sector_name, df):
    sector_companies = df[df['GICS Sector'] == sector_name]

    # Initialize data structure
    data = {
        'Company': [],
        'Ticker': [],
        'PE Ratio': [],
        'YoY Revenue Growth': [],
        'ROE': [],
        'Gross Margin': [],
        'R&D Spending': [],
        "Price to Book Value":[],
        "Return on Assets": [],
        "Debt to Equity":[]
    }

    # Fetch financial data for each company in the sector
    for ticker in sector_companies['Symbol']:
        stock = yf.Ticker(ticker)
        info = stock.info

        # Append company data
        data['Company'].append(info.get('longName', 'N/A'))
        data['Ticker'].append(ticker)
        data['PE Ratio'].append(info.get('forwardPE', 0))
        data['YoY Revenue Growth'].append(info.get('revenueGrowth', 0))
        data['ROE'].append(info.get('returnOnEquity', 0))
        data['Gross Margin'].append(info.get('grossMargins', 0))
        data['R&D Spending'].append(info.get('researchDevelopment', 0))
        data['Price to Book Value'].append(info.get('priceToBook', 0))
        data['Return on Assets'].append(info.get('returnOnAssets', 0))
        data['Debt to Equity'].append(info.get('debtToEquity', 0))

    # Convert to DataFrame
    df = pd.DataFrame(data)

    # Calculate averages
    averages = df.mean().to_dict()
    averages['Company'] = 'Average'
    averages['Ticker'] = ''
    df = df.append(averages, ignore_index=True)

    # Plot with Plotly
    fig = go.Figure(data=[go.Table(
        header=dict(values=list(df.columns), fill_color='paleturquoise', align='left'),
        cells=dict(values=[df[k].tolist() for k in df.columns], fill_color='lavender', align='left'))
    ])
    fig.show()

# Example: Plot for one sector, repeat for others as needed
plot_sector_data('Health Care', sp500_df)
```

```python
plot_sector_data('Energy', sp500_df)
```

```python
import pandas as pd
import yfinance as yf
import plotly.graph_objects as go

# Fetch S&P 500 list from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500_df = pd.read_html(url, header=0)[0]

# Filter for Information Technology companies
it_companies = sp500_df[sp500_df['GICS Sector'] == 'Information Technology']

# Group by GICS Sub-Industry
sub_industry_groups = it_companies.groupby('GICS Sub-Industry')

# Processing each Sub-Industry
for sub_industry, group in sub_industry_groups:
    print(f"Processing {sub_industry}...")
    sector_data = []

    for ticker in group['Symbol']:
        stock = yf.Ticker(ticker)
        info = stock.info

        company_data = {
            'Company': info.get('longName'),
            'Ticker': ticker,
            'PE Ratio': info.get('forwardPE'),
            'YoY Revenue Growth': info.get('revenueGrowth'),
            'ROE': info.get('returnOnEquity'),
            'Gross Margin': info.get('grossMargins'),
            'R&D Spending': info.get('researchDevelopment'),
            'PBV': info.get('priceToBook'),
            'ROA': info.get('returnOnAssets'),
            'Debt to Equity': info.get('debtToEquity'),
            # Simplified MACD indicator; in practice, you would calculate this from historical data
            'MACD Indicator': 'Not Calculated'  # Placeholder for simplicity
        }
        sector_data.append(company_data)

    df = pd.DataFrame(sector_data)
    # Creating a Plotly table for the sub-industry
    fig = go.Figure(data=[go.Table(
        header=dict(values=df.columns, fill_color='paleturquoise', align='left'),
        cells=dict(values=[df[col] for col in df.columns], fill_color='lavender', align='left'))
    ])
    fig.update_layout(title=f"{sub_industry} Sub-Industry")
    fig.show()
```

```python
# For each sector in the S&P 500
for sector in sp500_df['GICS Sector'].unique():
    sector_companies = sp500_df[sp500_df['GICS Sector'] == sector]
    growth_data = []

    for ticker in sector_companies['Symbol']:
        try:
            stock = yf.Ticker(ticker)
            info = stock.info

            growth_data.append({
                'Company': info.get('longName'),
            'Ticker': ticker,
            'PE Ratio': info.get('forwardPE'),
            'YoY Revenue Growth': info.get('revenueGrowth'),
            'ROE': info.get('returnOnEquity'),
            'Gross Margin': info.get('grossMargins'),
            'R&D Spending': info.get('researchDevelopment'),
            'PBV': info.get('priceToBook'),
            'ROA': info.get('returnOnAssets'),
            'Debt to Equity': info.get('debtToEquity'),
            # Simplified MACD indicator; in practice, you would calculate this from historical data
            'MACD Indicator': 'Not Calculated'
            })
        except Exception as e:
            print(f"Could not process {ticker}: {e}")

    # Sort by YoY Revenue Growth and select top 5
    df = pd.DataFrame(growth_data).sort_values(by='YoY Revenue Growth', ascending=False).head(5)

    # Create and display a Plotly table for top 5 companies in this sector
    fig = go.Figure(data=[go.Table(
        header=dict(values=df.columns, fill_color='paleturquoise', align='left'),
        cells=dict(values=[df[col] for col in df.columns], fill_color='lavender', align='left'))
    ])
    fig.update_layout(title=f"Top 5 Companies by YoY Revenue Growth in {sector} Sector")
    fig.show()
```
